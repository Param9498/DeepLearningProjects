```python
g = open('reviews.txt', 'r')
reviews = list(map(lambda x:x[:-1], g.readlines()))
g.close()
```

```python
g = open('labels.txt', 'r')
labels = list(map(lambda x:x[:-1].upper(), g.readlines()))
g.close()
```

```python
vocab = set()

for review in training_reviews:
    for word in review.split(" "):
        vocab.add(word)
        
vocab
```

```python
word2vector = {}

for i, word in enumerate(vocab):
    word2vector[word] = i
    
word2vector
```

```python
import numpy as np

training_list = np.zeros((len(training_reviews), len(word2vector)))
training_labels_list = np.zeros((len(training_labels), 1))


for i in range(len(training_reviews)):
    for word in training_reviews[i].split(" "):
        if(word in word2vector):
            training_list[i][word2vector[word]] = 1
            if(training_labels[i] == 'POSITIVE'):
                training_labels_list[i][0] = 1
            else:
                training_labels_list[i][0] = 0

```

```python
print(training_reviews[2])
print(training_labels[2])
print(training_labels_list[2])
for index, e in enumerate(training_list[2]):
    if(e == 1):
        for word, key in word2vector.items():
            if key == index:
                print(word)
```

```python
assert(len(training_list) == len(training_reviews))
```

```python
import numpy as np
from keras.utils import np_utils
import tensorflow as tf

from keras.models import Sequential
from keras.layers.core import Dense, Activation, Flatten, Dropout

training_list = training_list
training_labels_list = training_labels_list

training_list[0]
```

```python
training_labels_list[0]
```

```python
model = Sequential()
model.add(Dense(10, input_shape=(len(word2vector), )))
model.add(Dense(1))
model.add(Activation("sigmoid"))

model.compile(loss="mean_squared_error", optimizer="sgd", metrics = ['accuracy'])

model.summary()

history = model.fit(training_list, training_labels_list, epochs=10, batch_size=100, verbose=2)
```
