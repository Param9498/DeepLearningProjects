```python
g = open('reviews.txt', 'r')
reviews = list(map(lambda x:x[:-1], g.readlines()))
g.close()
```

```python
g = open('labels.txt', 'r')
labels = list(map(lambda x:x[:-1].upper(), g.readlines()))
g.close()
```

```python
training_reviews = reviews[:-1000]
training_labels = labels[:-1000]
```

```python
vocab = set()

for review in training_reviews:
    for word in review.split(" "):
        vocab.add(word)
        
vocab
```

```python
word2vector = {}

for i, word in enumerate(vocab):
    word2vector[word] = i
    
word2vector
```

```python
len(word2vector)
```

```python
list(word2vector.keys())[1]
```

```python
import numpy as np

training_list = np.zeros((len(training_reviews), len(word2vector)))
training_labels_list = np.zeros((len(training_labels)))


for i in range(len(training_reviews)):
    for word in training_reviews[i].split(" "):
        if(word in word2vector):
            training_list[i][word2vector[word]] = 1
            if(training_labels[i] == 'POSITIVE'):
                training_labels_list[i] = 1
            else:
                training_labels_list[i] = 0

```

```python
print(training_reviews[1])
print(training_labels[1])
print(training_labels_list[1])
for index, e in enumerate(training_list[1]):
    if(e == 1):
        for word, key in word2vector.items():
            if key == index:
                print(word)
```

```python
assert(len(training_list) == len(training_reviews))
```
