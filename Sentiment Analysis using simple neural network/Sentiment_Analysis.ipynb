# We see that words like "the" occour in both positive and negative reviews almost the same number of times

#### So let's find out the ratio : number of times a word occurs in positive / number of times a word occurs in negative

#### This will make all the words that :
    occur similar number of times in the positive and negative reviews almost equal to 1
    occur more in positive reviews and less in negative reviews greater than 1
    occur less in positive reviews and more in negative reviews less than 1 and greater than 0

```python
pos_neg_ratios = Counter()
for word, count in list(total_counts.most_common()):
    if(count > 100):
        pos_neg_ratios[word] = positive_counts[word] / float(negative_counts[word] + 1)
```

```python
print("Pos-to-neg ratio for ' ' = " + str(pos_neg_ratios[""]))
```

## Convert ratios to logs so that the values:
    closer to 0 are negated
    greater than 1 are positive
    almost equal to 1 are made 0

```python
for word, ratio in pos_neg_ratios.most_common():
    if(ratio > 1):
        pos_neg_ratios[word] = np.log(ratio)
    else:
        pos_neg_ratios[word] = - np.log((1/(ratio + 0.01)))
```

```python
pos_neg_ratios.most_common()
```

```python
print("Pos-to-neg ratio for 'the' = {}".format(pos_neg_ratios["the"]))
```

## Building the Sentiment Analysis class

```python
import time
import sys
import numpy as np
from collections import Counter

class SentimentNetwork:
    def __init__(self, reviews, labels, hidden_nodes, learning_rate):
        np.random.seed(1)
        
        # self.load_data()
        
        self.pre_process_data(reviews, labels)
        
        self.init_network(hidden_nodes, 1, learning_rate)
        
    def load_data(self):
        g = open('reviews.txt', 'r')
        self.reviews = list(map(lambda x : x[:-1],g.readlines()))
        g.close()
        g = open('labels.txt', 'r')
        self.labels = list(map(lambda x : x[:-1].upper(), g.readlines()))
        g.close()
        
    def pre_process_data(self, reviews, labels):
        review_vocab = set()
        
        for review in reviews:
            for word in review.split(" "):
                review_vocab.add(word)
        
        self.review_vocab = list(review_vocab)
        
        label_vocab = set()
        
        for label in labels:
            label_vocab.add(label)
            
        self.label_vocab = list(label_vocab)
        
        self.review_vocab_size = len(self.review_vocab)
        
        self.label_vocab_size = len(self.label_vocab)
        
        
        self.word2index = {}
        
        for i, word in enumerate(self.review_vocab):
            self.word2index[word] = i
            
        self.label2index = {}
        
        for i, label in enumerate(self.label_vocab):
            self.label2index[label] = i
            
    def init_network(self, hidden_nodes, output_nodes, learning_rate):
        
        self.input_nodes = len(self.review_vocab)
        
        self.output_nodes = 1
        
        self.hidden_nodes = hidden_nodes
        
        self.learning_rate = learning_rate
        
        # We are gonna be making a neural network with only one hidden layer. SO there will be two weight matrices that
        # we will need. One is weights_0_1[input_nodes][hidden_nodes] which is a matrix of weights between the input and the hidden layer. Other 
        # is weights_1_2[hidden_nodes][output_nodes] which will be the weights between the hidden layer and the output layer.
        
        self.weights_0_1 = np.zeros((self.input_nodes, self.hidden_nodes))
        
        self.weights_1_2 = np.random.normal(0.0, self.output_nodes**-0.5, (self.hidden_nodes, self.output_nodes))
        
        # Now lets create the input layer
        
        self.layer_0 = np.zeros((1, self.input_nodes))
        
    def update_input_layer(self, review, label, positive_counts, negative_counts, total_counts):
        
        self.layer_0 *= 0
        if(label == "POSITIVE"):
            for word in review.split(" "):
                if(word in self.word2index.keys()):
                    if(self.layer_0[0][self.word2index[word]] == 0):
                        self.layer_0[0][self.word2index[word]] = 1
                    else:
                        self.layer_0[0][self.word2index[word]] += 1
                        self.layer_0[0][self.word2index[word]] = self.layer_0[0][self.word2index[word]] / (positive_counts[word] + 1)
        else:
            for word in review.split(" "):
                if(word in self.word2index.keys()):
                    if(self.layer_0[0][self.word2index[word]] == 0):
                        self.layer_0[0][self.word2index[word]] = 1
                    else:
                        self.layer_0[0][self.word2index[word]] += 1
                        self.layer_0[0][self.word2index[word]] = self.layer_0[0][self.word2index[word]] / (negative_counts[word] + 1)
                    
                
    def get_target_label(self, label):
        if(label == "POSITIVE"):
            return 1
        else:
            return 0
        
    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))
    
    def sigmoid_output_to_derivative(self, output):
        return output * (1 - output)
    
    def train(self, training_reviews, training_labels):
        assert(len(training_reviews) == len(training_labels))
        
        correct_so_far = 0
        
        start = time.time()
        
        positive_counts = Counter()
        negative_counts = Counter()
        total_counts = Counter()
        
        for i in range(len(training_reviews)):
            if(training_labels[i] == "POSITIVE"):
                for word in training_reviews[i].split(" "):
                    positive_counts[word] += 1
                    total_counts[word] += 1
            elif(training_labels[i] == "NEGATIVE"):
                for word in training_reviews[i].split(" "):
                    negative_counts[word] += 1
                    total_counts[word] += 1
        
        for i in range(len(training_reviews)):
            review = training_reviews[i]
            label = training_labels[i]
            
            self.update_input_layer(review, label, positive_counts, negative_counts, total_counts)
            
            x1 = self.layer_0
            y1 = x1
            
            x2 = y1.dot(self.weights_0_1)
            y2 = x2
            
            x3 = y2.dot(self.weights_1_2)
            y3 = self.sigmoid(x3)
            
            
            delta_3 = (y3 - self.get_target_label(label)) * self.sigmoid_output_to_derivative(y3)
            self.weights_1_2 -= self.learning_rate * y2.T.dot(delta_3)
            
            delta_2 = delta_3.dot(self.weights_1_2.T) * 1
            self.weights_0_1 -= self.learning_rate * y1.T.dot(delta_2)
            
            if(y3 >= 0.5 and label == 'POSITIVE'):
                correct_so_far += 1
            elif(y3 < 0.5 and label == 'NEGATIVE'):
                correct_so_far += 1
                
            elapsed_time = (time.time() - start)
            reviews_per_second = i / elapsed_time if elapsed_time > 0 else 0
            
            sys.stdout.write("\rProgress:" + str(100 * i/float(len(training_reviews)))[:4] \
                             + "% Speed(reviews/sec):" + str(reviews_per_second)[0:5] \
                             + " #Correct:" + str(correct_so_far) + " #Trained:" + str(i+1) \
                             + " Training Accuracy:" + str(correct_so_far * 100 / float(i+1))[:4] + "%")
            if(i % 2500 == 0):
                print("")
                
    def test(self, testing_reviews, testing_labels):
        """
        Attempts to predict the labels for the given testing_reviews,
        and uses the test_labels to calculate the accuracy of those predictions.
        """
        
        # keep track of how many correct predictions we make
        correct = 0

        # we'll time how many predictions per second we make
        start = time.time()

        # Loop through each of the given reviews and call run to predict
        # its label. 
        for i in range(len(testing_reviews)):
            pred = self.run(testing_reviews[i], testing_labels[i])
            if(pred == testing_labels[i]):
                correct += 1
            
            # For debug purposes, print out our prediction accuracy and speed 
            # throughout the prediction process. 

            elapsed_time = float(time.time() - start)
            reviews_per_second = i / elapsed_time if elapsed_time > 0 else 0
            
            sys.stdout.write("\rProgress:" + str(100 * i/float(len(testing_reviews)))[:4] \
                             + "% Speed(reviews/sec):" + str(reviews_per_second)[0:5] \
                             + " #Correct:" + str(correct) + " #Tested:" + str(i+1) \
                             + " Testing Accuracy:" + str(correct * 100 / float(i+1))[:4] + "%")
    
    def run(self, review, label):
        """
        Returns a POSITIVE or NEGATIVE prediction for the given review.
        """
        # TODO: Run a forward pass through the network, like you did in the
        #       "train" function. That means use the given review to 
        #       update the input layer, then calculate values for the hidden layer,
        #       and finally calculate the output layer.
        #
        #       Note: The review passed into this function for prediction 
        #             might come from anywhere, so you should convert it 
        #             to lower case prior to using it.
        
        positive_counts = Counter()
        negative_counts = Counter()
        total_counts = Counter()
        
        
        if(label == "POSITIVE"):
            for word in review.split(" "):
                positive_counts[word] += 1
                total_counts[word] += 1
        elif(label == "NEGATIVE"):
            for word in review.split(" "):
                negative_counts[word] += 1
                total_counts[word] += 1
                    
        self.update_input_layer(review, label, positive_counts, negative_counts, total_counts)
            
        x1 = self.layer_0
        y1 = x1

        x2 = y1.dot(self.weights_0_1)
        y2 = x2

        x3 = y2.dot(self.weights_1_2)
        y3 = self.sigmoid(x3)
        
        # TODO: The output layer should now contain a prediction. 
        #       Return `POSITIVE` for predictions greater-than-or-equal-to `0.5`, 
        #       and `NEGATIVE` otherwise.
        if(y3 >= 0.5):
            return "POSITIVE"
        else:
            return "NEGATIVE"
        
```

```python
mlp = SentimentNetwork(reviews[:-1000],labels[:-1000], 10, learning_rate=0.001)
mlp.train(reviews[:-1000],labels[:-1000])
```
